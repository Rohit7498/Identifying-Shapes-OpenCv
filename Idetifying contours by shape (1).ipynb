```python
# MY PROJECT : Idetifying contours by shape
import cv2
import numpy as np

# First we load the image on which shapes has to be identified
image = cv2.imread('images/someshapes.jpg')
cv2.imshow('Identifying Shapes',image)
cv2.waitKey(0)

# We convert it into grayscale finding contours
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(image_gray, 127, 255, 1)

# We now find contours of the original image
im, contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE) # Retrieval mode is lis i.e all 

# Now we iterate over the contour array and find the approx polygon 
for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt, True), True) # It returns the approximate polygon 0.01 approx can be 
                                                                        # modified accordingly
    if len(approx) == 3:
        shape_name = 'Triangle'
        cv2.drawContours(image, [cnt], 0, (0,255,0), -1) # We fill the triangle image with green
        
        # we find the contour centroid just for the sake of putting text in middle
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        
        cv2.putText(image, shape_name, (cx-65,cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
        
    elif len(approx) == 4:
        # Since 4 vertex shape can be square or rectange so we find the height and width to sort
        
        (x, y, w, h) = cv2.boundingRect(cnt)
        
        if abs(w-h) <= 3:
            shape_name = 'Square'
            cv2.drawContours(image, [cnt], 0, (255,0,0), -1) # We fill the square image with blue
        
        # we find the contour centroid just for the sake of putting text in middle
            M = cv2.moments(cnt)
            cx = int(M['m10'] / M['m00'])
            cy = int(M['m01'] / M['m00'])
        
            cv2.putText(image, shape_name, (cx-60,cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
            
        else:
            shape_name = 'Rectangle'
            cv2.drawContours(image, [cnt], 0, (0,0,255), -1) # We fill the Rectangle image with red
        
        # we find the contour centroid just for the sake of putting text in middle
            M = cv2.moments(cnt)
            cx = int(M['m10'] / M['m00'])
            cy = int(M['m01'] / M['m00'])
        
            cv2.putText(image, shape_name, (cx-70,cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
            
    elif len(approx) == 10:
        
        shape_name = 'Star'
        cv2.drawContours(image, [cnt], 0, (0,255,255), -1) # We fill the Star image with Yellow
        
        # we find the contour centroid just for the sake of putting text in middle
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        
        cv2.putText(image, shape_name, (cx-50,cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)

    elif len(approx) >= 15:
        
        shape_name = 'Circle'
        cv2.drawContours(image, [cnt], 0, (120,200,50), -1)
        
        # we find the contour centroid just for the sake of putting text in middle
        M = cv2.moments(cnt)
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        
        cv2.putText(image, shape_name, (cx-60,cy), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0), 1)
        
    # We Display identified images one by one
    
    cv2.imshow('Identifying Shapes', image)
    cv2.waitKey(0)
    
cv2.destroyAllWindows()
```
